```python
import pandas as pd
from dotenv import load_dotenv
import os

load_dotenv()

storage_options = {
    "key": os.getenv("AWS_ACCESS_KEY"),
    "secret": os.getenv("AWS_SECRET_KEY"),
}

path = (
    "s3://nyc-taxi-analytics-dibya/"
    "silver/trips/"
    "pickup_date=2023-01-01/"
)

df = pd.read_parquet(
    path,
    storage_options=storage_options,
)

print(df.head())
print(df.shape)
```

```python
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()

storage_option={
    "key":os.environ.get("AWS_ACCESS_KEY"),
    "secret":os.environ.get("AWS_SECRET_KEY"),
}

path = (
    "s3://nyc-taxi-analytics-dibya/"
    "silver/trips/"
    "pickup_date=2023-01-01/"
)

df= pd.read_parquet(
    path,
    storage_options=storage_option,
)

print(df.shape)
```
